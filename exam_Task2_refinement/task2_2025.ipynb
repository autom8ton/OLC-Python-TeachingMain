```python
# Dictionary
# key and value pair

capital_cities = {"malaysia": "kuala lumpur",
                  "germany":"berlin",
                  "thailand":"bangkok"}

# to retrieve value
cap1 = capital_cities["germany"]  # retrieving the value of bangkok
print(cap1)

# change the value of a key
capital_cities["thailand"] = "phuket"
print(capital_cities)

# same syntax for changing and for adding
capital_cities["vietnam"] = "ho chi minh city"
capital_cities["vietnam"] = "hanoi"
print(capital_cities)

# loop through the key and values of the dictionary
for thekey, thevalue in capital_cities.items(): # retrieve both key and value
    print(f"Country: {thekey} Capital: {thevalue}")

```
